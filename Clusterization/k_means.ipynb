# K-Means Clusterization

## Imports

```python
import pandas as pd
import numpy  as np

from matplotlib import pyplot  as plt
from sklearn    import cluster as ct
from sklearn    import metrics as mt
```

## Loading Dataset

```python
ds = pd.read_csv('X_dataset.csv')
```

## Training Model

```python
n_clusters = np.arange(2, 51, 1)

ss_list = []
for i in n_clusters:
    # Define
    kmeans = ct.KMeans(n_clusters=i, max_iter=1000)

    # training
    labels = kmeans.fit_predict(ds)

    # performance
    ss_avg = mt.silhouette_score(ds, labels)
    ss_list.append(ss_avg)

max_ss = ss_list.index(max(ss_list))
print(f'For n_clusters= {n_clusters[max_ss]}, the average silhoutte is {max(ss_list):.2f}')
```

```python
plt.plot(n_clusters, ss_list, marker='o')
plt.xlabel('Clusters K')
plt.ylabel('Silhouette Score Average')
```
