# Affinity Propagation

## Imports

```python
import pandas as pd
import numpy  as np
import time

from IPython    import display
from matplotlib import pyplot   as plt
from sklearn    import cluster  as c
from sklearn    import metrics  as mt
```

## Loading Dataset

```python
ds = pd.read_csv('X_dataset.csv')
```

## Training

```python
# testing parameters
p = np.arange( -1, -51, -1)

ss_list = []
for i in range(len(p)):
    print(f'Preference values: {p[i]}')
    
    # model definition
    model = c.AffinityPropagation(preference=p[i], max_iter=1000)
    
    # model training
    model.fit(ds)
    
    # predict (clustering)
    labels = model.predict(ds)

    # performance
    ss = mt.silhouette_score(ds, labels)
    ss_list.append(ss)

    # plot
    plt.plot(p[:i], ss_list[:i])

    plt.title('Affinity Propagation')
    plt.xlabel('Preference')
    plt.ylabel('Silhouette Score')

    display.clear_output(wait=False)
    plt.show()
    plt.gcf()
    time.sleep(0.05)
```

```python
best_p = ss_list.index(max(ss_list))

print(f'Number of clusters: {len(np.unique(labels))}')
print(f'Silhouette Score: {max(ss_list):.3f}')
```
